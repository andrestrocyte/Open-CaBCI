```python
import matplotlib
%matplotlib qt

import matplotlib.pyplot as plt
# to run GUI event loop
plt.ion()
#%matplotlib tk
%autosave 180
#%load_ext autoreload
#%autoreload 2
import numpy as np
import nidaqmx

#from nidaqmx.constants import LineGrouping
#from nidaqmx.constants import Edge
#from nidaqmx.constants import AcquisitionType 
from nidaqmx import stream_writers

import time
```

```python
############################################
############################################
############################################
#  Initialize speaker task;
water_Task = nidaqmx.Task()

# 
water_Task.ao_channels.add_ao_voltage_chan('Dev3/ao1')
water_Task.timing.cfg_samp_clk_timing(rate = 1000, 
                                     #samps_per_chan=100,  # in continuos mode this is the buffer
                                     sample_mode=nidaqmx.constants.AcquisitionType.CONTINUOUS)
                                     #sample_mode=nidaqmx.constants.AcquisitionType.FINITE)

# 
water_Writer = nidaqmx.stream_writers.AnalogSingleChannelWriter(water_Task.out_stream, 
                                                                auto_start=True)

# realtime streaming with callback
start = time.time()
ctr=0

duration = 10000  # time in microsceconds
for k in range(20):
    #pulse = get_TTL_pluse(duration)
    for p in range(duration):
        water_Writer.write_one_sample(5)
    
    # pause and give 0 pluse
    for q in range(100000-duration):
        water_Writer.write_one_sample(0)
    #write_many_sample(pulse)

    #time.sleep(1)
    print ("pulse: ", k, " duration: ", duration/1E6, "sec")
#    
water_Task.stop()
water_Task.close()

print ("DONE")
```
